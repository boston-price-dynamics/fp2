```python
import pandas as pd
import geopandas as gpd

df = pd.read_csv('assessment_2024.csv', dtype='str')
df
```

```python
geo_df = gpd.read_file("parcels_2023.geojson")
geo_df
```

```python
geo_df["lon"] = geo_df["geometry"].centroid.x
geo_df["lat"] = geo_df["geometry"].centroid.y
geo_df
```

```python
len(df['GIS_ID'].unique())
```

```python
geo_df['MAP_PAR_ID'] = geo_df['MAP_PAR_ID'].astype(str)
```

```python
len(geo_df['MAP_PAR_ID'].unique())
```

```python
df
```

```python
merged_df = pd.merge(df, geo_df, how='inner', left_on='GIS_ID', right_on='MAP_PAR_ID')
merged_df
```

```python
output_df = merged_df[['GIS_ID', 'ST_NUM', 'ST_NAME', 'UNIT_NUM', 'CITY', 'ZIP_CODE', 'LU', 'BLDG_TYPE', 'OWN_OCC', 'TOTAL_VALUE', 'YR_BUILT', 'lon', 'lat']]
output_df
```

```python
output_df = output_df[output_df['LU'].isin({'A', 'CD', 'R1', 'R2', 'R3', 'R4'})]
output_df = output_df.dropna(subset=['TOTAL_VALUE', 'YR_BUILT'])
output_df
```

```python
output_df['TOTAL_VALUE'] = [a.replace(',', '') for a in output_df['TOTAL_VALUE']]
# output_df['TOTAL_VALUE'] = output_df['TOTAL_VALUE'].astype(int)
output_df.to_csv('../assess_lat_long.csv')
```
