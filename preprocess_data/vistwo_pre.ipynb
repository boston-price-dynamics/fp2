```python
all_results = []
for yr in range(2010, 2025):
    df = pd.read_csv(f"property-assessment-fy{yr}.csv", dtype="str")
    df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
    df = df[df['ST_NAME'].str.contains('GERMAIN', case=False, na=False) | df['ST_NAME'].str.contains('BOTOLPH', case=False, na=False)]

    if 'ST_NAME_SUF' in df.columns:
        df['ST_NAME'] = df['ST_NAME'] + ' ' + df['ST_NAME_SUF']
    if 'AV_TOTAL' in df.columns:
        df['TOTAL_VALUE'] = df['AV_TOTAL']
    if 'PTYPE' in df.columns:
        df['LUC'] = df['PTYPE']
    
    df['ST_NAME'] = df['ST_NAME'].str.upper()
    
    df = df[['ST_NUM', 'ST_NAME', 'LUC', 'LU', 'TOTAL_VALUE']].dropna()
    if yr == 2021:
        df['TOTAL_VALUE'] = df['TOTAL_VALUE'].astype(str).apply(lambda x: x.replace('$', '').replace(',', '')[:-3])
    elif yr > 2021:
        df['TOTAL_VALUE'] = df['TOTAL_VALUE'].astype(str).apply(lambda x: x.replace('$', '').replace(',', ''))
    df = pd.merge(rel_df, df, on=['ST_NUM', 'ST_NAME'], how='left')
    df['TOTAL_VALUE'] = pd.to_numeric(df['TOTAL_VALUE'])  # Convert to numeric type if needed
    result_df = df.groupby(['ST_NUM', 'ST_NAME']).agg({'TOTAL_VALUE': 'max'}).reset_index()
    print(yr, len(result_df.dropna()))
    result_df['yr'] = yr
    result_df = pd.merge(rel_df, result_df, on=['ST_NUM', 'ST_NAME'], how='left')
    if yr > 2010:
        old_values = []
        for i, r in result_df.iterrows():
            old_value = int(first_df[(first_df['ST_NUM'] == r['ST_NUM']) & (first_df['ST_NAME'] == r['ST_NAME'])].iloc[0]['TOTAL_VALUE'])
            old_values.append(old_value)
        

        result_df['old_val'] = old_values
        result_df['PercentChg'] = (result_df['TOTAL_VALUE'] - result_df['old_val']) / result_df['old_val']
        result_df = result_df.drop(['old_val'], axis=1)

    else:
        result_df['PercentChg'] = 0

    if yr == 2010:
        first_df = result_df
    all_results.append(result_df)
```

```python
all_results_df = pd.concat(all_results)
all_results_df
```
